**Question 2 (5 points): Functions and Conditionals**

A small business offers a discount to its customers based on their total purchase amount. Write a Python function called `calculate_discount` that takes the total purchase amount as a parameter and returns the discount amount based on the following rules:

- If the purchase amount is less than $50, no discount is applied.

- If the purchase amount is between $50 and 100 (inclusive), apply a 5% discount.

- If the purchase amount is greater than $100, apply a 10% discount.

Test your function by calculating the discount for a purchase of $75 and 120.


````python
def calculate_discount(total_purchase):
    if total_purchase < 50:
        discount_amount = 0.0
    elif total_purchase <= 100:
        discount_amount = total_purchase * 0.05
    else:
        discount_amount = total_purchase * 0.10
    return discount_amount

amount1 = 75
amount2 = 120

discount1 = calculate_discount(amount1)
discount2 = calculate_discount(amount2)

print(f"Purchase: ${amount1} | Discount: ${discount1}")
print(f"Purchase: ${amount2} | Discount: ${discount2}")
    
````

**Question 3 (5 points): Iterations and Loops**

You are given a list of monthly sales figures for a small business for the past year:

```python
sales = [25000, 28000, 32000, 34000, 38000, 41000, 45000, 48000, 50000, 52000, 55000, 60000]
```

Write a Python program that calculates and displays the average monthly sales and the months (indices) where the sales exceeded the average.

You may calculate the average manually, or with another function from pandas, numpy, etc. if you are familiar with it.

Hint: Here's an example output you can use to model your code. The months and numbers are NOT the correct answer!!

<pre>
Average Monthly Sales: $42333.33 
Months with Sales Exceeding Average:
Month 4 : $25000.00
Month 5 : $43000.00

Month 12 : $10000.00
</pre>

````python
sales = [25000, 28000, 32000, 34000, 38000, 41000, 45000, 48000, 50000, 52000, 55000, 60000]

total_sales = sum(sales)
count_months = len(sales)
average_sales = total_sales / count_months

print(f"Average Monthly Sales: ${average_sales:.2f}")
print("Months with Sales Exceeding Average:")

for i in range(len(sales)):
    current_sale = sales[i]

    if current_sale > average_sales:
        print(f"Month {i + 1} : ${current_sale:.2f}")
````

```python
orders = {
    "Order001": {"product": "Widget A", "quantity": 5},

    "Order003": {"product": "Widget A", "quantity": 8},
    "Order004": {"product": "Widget C", "quantity": 3}
}
```

Write a Python program that displays the total quantity of products ordered per Order #.
Here's an example (this is NOT the right answer! just the formatting)

**CHALLENGE! +5 extra credit 😎**
Write a Python program that calculates the total quantity of each product ordered and displays the results. Additionally, identify and display the product with the highest total quantity ordered.


````python
orders = {
    "Order001": {"product": "Widget A", "quantity": 5},
    "Order002": {"product": "Widget B", "quantity": 10},
    "Order003": {"product": "Widget A", "quantity": 8},
    "Order004": {"product": "Widget C", "quantity": 3}
}

for order_id, details in orders.items():
    print(f"{order_id}: {details['quantity']} items")

product_totals = {}

for details in orders.values():
    product = details['product']
    qty = details['quantity']

    product_totals[product] = product_totals.get(product, 0) + qty

print("\nTotal Quantity per Product:")
for product, total in product_totals.items():
    print(f"{product}: {total}")


max_product = None
max_qty = 0

for product, total in product_totals.items():
    # Compare each total to the current maximum
    if total > max_qty:
        max_qty = total
        max_product = product

print(f"\nProduct with highest total quantity: {max_product} ({max_qty} items)")
````
